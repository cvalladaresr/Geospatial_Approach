# Descriptive Statistics
The purpose of this notebook is to perform an exploratory analysis of the data in order to facilitate the interpretation of results and learn more about the phenomenon under study.

## Incorporated Variables

- **grid_id**: Unique identifier of the spatial quadrant (grid) where the observations are grouped. Each grid_id corresponds to a geographic polygon defined in the 1x1 km grid system within the Basque Country. To consult the reference polygon for each grid, the file *grilla_1km_pais_vasco.geojson* was created. We have 7,718 grids.

- **year**: Year of the observation or measurement. We will work with a series from 2018 to 2022.
- **month**: Month of the observation (numeric format, 1 = January, 12 = December).

- **NO2**: Monthly average concentration of nitrogen dioxide (NO₂) in the corresponding grid. Unit in mol/m² or similar, depending on the satellite source.

- **CO**: Monthly average concentration of carbon monoxide (CO). May contain null values ​​(NaN) if no data is available.

- **SO₂**: Monthly average concentration of sulfur dioxide (SO₂). May contain NaN values.

- **O₃**: Monthly average concentration of tropospheric ozone (O₃). May contain NaN values.

- **AAI**: Aerosol Absorbing Index — indicator of absorbent aerosols (such as dust or smoke) detected in the atmosphere. Negative values ​​may indicate clear conditions.

- **NTL**: Nighttime Lights: average level of nighttime luminance (derived from satellite imagery). Indicates human activity and infrastructure.

- **LST**: Land Surface Temperature: temperature of the Earth's surface (in °K). To convert to °C, multiply by 0.02.
- **PM25**: Estimated concentration of fine particulate matter (PM2.5). Unit in mg/m2.

- **has_plant**: Binary indicator: 1 if there is at least one steel plant within the grid, 0 otherwise. Derived from a spatial join between plant locations and polygons.

- **production**: Total industrial production (sum) corresponding to the grid in that month and year. Calculated by adding the values ​​of all plants located within the grid_id. If it is NaN, no plants were reported in that month/grid.

## Additional Variables
During the process of creating this notebook, additional fields necessary before modeling were added.

- **SO2_imp1**: This is the same **SO2** series but imputed for NaN data. This variable uses interpolation for imputation.

- **SO2_imp2**: This is the same **SO2** series but imputed for NaN data. In this variable, they used an *extrapolation* using the month-over-month rates in OMI for imputation.

- **LST_imp**: This is the same **LST** series but imputed for Nan data. In this variable, they used an *interpolation* for imputation.

- **exclude**: A binary variable that signals the observations that should be excluded due to missing data in NTL across the entire series.

## Creación de variables *lag*
Se crean ya que emisiones del pasado pueden explicar la producción presente.

```python
# Primero, ordenar por grid_id y fecha para asegurar orden correcto
df_lag = df.sort_values(['grid_id', 'fecha'])

# Lista de contaminantes a los que quieres aplicar el lag
contaminantes = ['NO2', 'CO', 'SO2', 'O3', 'AAI', 'NTL', 'LST', 'PM25']

# Crear lags de 1, 2 y 3 meses para cada contaminante por grid_id
for cont in contaminantes:
    for lag in [1, 2, 3]:
        df_lag[f'{cont}_lag_{lag}'] = df_lag.groupby('grid_id')[cont].shift(lag)
```

```python
df_lag.head()
```

```python
# Calcular matriz de correlación
contaminantes = ['NO2', 'CO', 'SO2', 'O3', 'AAI', 'NTL', 'LST', 'PM25']
lags = [1, 2, 3]

# Generar lista de todas las columnas (originales + lags)
contaminantes_lags = contaminantes + [f'{c}_lag_{l}' for c in contaminantes for l in lags]

# Calcular matriz de correlación entre todas esas columnas
corr_matrix = df_lag[contaminantes_lags].corr()

# Mostrar heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0, annot=False)
plt.title('Matriz de correlación entre contaminantes y sus lags')
plt.tight_layout()
plt.show()
```

```python
# Asegúrate de excluir filas donde 'produccion' sea NaN
df_corr = df_lag[contaminantes_lags + ['produccion']].dropna(subset=['produccion'])

# Calcular correlación de cada variable (original o lag) con produccion
corr_produccion = df_corr.corr()['produccion'].drop('produccion')

# Mostrar gráfico
plt.figure(figsize=(10, 6))
corr_produccion.sort_values().plot(kind='barh', color='steelblue')
plt.title('Correlación entre producción y contaminantes (incluyendo lags)')
plt.xlabel('Correlación')
plt.grid(True, axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Imputación a SO2

Como alternativa de imputación, creamos una regresión de SO2 respecto a todos los otros contaminantes y sus Lags.
  
**Regresión lineal R-square = 0.24**: Es muy bajo, así que se descarta esta alternativa.  
**Random Forest lineal R-square = 0.55**: Si bien mejora, sigue siendo bajo.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
import pandas as pd

# Definir contaminantes excepto SO2
contaminantes = ['NO2', 'CO', 'O3', 'AAI', 'NTL', 'LST', 'PM25']
lags = [1, 2, 3]

# Crear lista de predictores (sin incluir SO2 ni sus lags)
predictors = contaminantes + [f'{c}_lag_{l}' for c in contaminantes for l in lags]

# Eliminar filas con NaN en variables relevantes
df_model = df[['SO2'] + predictors].dropna()

# Separar X e y
X = df_model[predictors]
y = df_model['SO2']

# Estandarizar X
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Entrenar/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Modelo lineal
model = LinearRegression()
model.fit(X_train, y_train)

# Predicciones y métricas
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Mostrar métricas
print(f'R² del modelo: {r2:.3f}')
print(f'RMSE: {rmse:.5f}')

# Coeficientes (ordenados por magnitud)
coef_df = pd.DataFrame({
    'Variable': predictors,
    'Coeficiente': model.coef_
}).sort_values(by='Coeficiente', key=abs, ascending=False)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

r2_score(y_test, y_pred)
```

### Imputación 1:  con interpolación  
Se decide guardar dos versiones de imputación. La primera será dejar intactos todos los meses en que tenemos medición de SO2 en Copernicus, y para los meses que no tenemos medición realizar una interpolación con promedios del nivel anterior y posterior.
  
La razón de está imputación está en el archivo: *SO2 copernicus vs OMI.xlsx* en la hoja: *interpolado v1*. En resumen:  El producto OMI/Aura es una excelente alternativa para imputar valores faltantes de SO₂, especialmente en fechas donde Sentinel-5P no tiene cobertura. Aunque la resolución espacial de OMI (13×24 km) es más baja que la de Sentinel-5P (7×3.5 km o mejor), sigue siendo útil para llenar huecos o hacer análisis complementarios. Lo principal que observamos es la estacionalidad marcada donde justo los meses que nos faltan (diciembre y enero) son los de mayor producción.

Antes de interpolar, mencionar que **todo dato original de Sentinel 5P se deja y solo agregamos la interpolación para los meses missing a nivel de grid**.

La interpolación es a nivel de mes, año y grid.

```python
df.head()
```

```python
# Asegurarnos que df esté ordenado
df = df.sort_values(["grid_id", "year", "month"]).copy()

# Inicializar columna imputada
df["SO2_imp1"] = df["SO2"]

# Función para imputar por grid_id sin extrapolar fuera de los datos conocidos
def imputar_so2(grupo):
    grupo = grupo.set_index("fecha")
    grupo["SO2_imp1"] = grupo["SO2"].interpolate(method="linear", limit_area="inside")
    return grupo.reset_index()

# Aplicar por cada grid
df = df.groupby("grid_id", group_keys=False).apply(imputar_so2)

# Orden final
df = df.sort_values(["grid_id", "fecha"])
```

```python
# 1. Agrupar por año y mes
df_grouped = df.groupby(["year", "month"], as_index=False)[["SO2_imp1", "SO2_OMI"]].mean()

# 2. Columna de fecha (primer día de cada mes)
df_grouped["fecha"] = pd.to_datetime(df_grouped[["year", "month"]].assign(day=1))

# 3. Crear figura y ejes
fig, ax1 = plt.subplots(figsize=(12,6))

# Ancho de barra: 20 días (para que se vea bien en Colab)
bar_width = pd.Timedelta(days=20)

# Eje 1 - Barras SO2
ax1.bar(df_grouped["fecha"], df_grouped["SO2_imp1"],
        width=bar_width, color="green", label="SO2 interpolado")
ax1.set_ylabel("SO2_imp1", color="green")
ax1.tick_params(axis="y", labelcolor="green")

# Eje 2 - Línea SO2_OMI
ax2 = ax1.twinx()
ax2.plot(df_grouped["fecha"], df_grouped["SO2_OMI"],
         color="red", marker="o", label="SO2_OMI")
ax2.set_ylabel("SO2_OMI", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# Título y formato
fig.suptitle("Comparación mensual SO2 imputación 1 (barra) vs SO2_OMI (línea)", fontsize=14)
fig.autofmt_xdate()

# Leyendas
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.show()
```

### Imputación 2: extrapolación
En esta versión de la imputación también mantenemos lo niveles originales de SO2 de Sentinel 5P.
Para imputar missings, se utiliza la tasa de variación intermensual en OMI para extrapolar el mes anterior.

```python
# Partimos de df ordenado
df = df.sort_values(["grid_id", "year", "month"]).copy()

# Creamos SO2_imp2 inicialmente igual a SO2
df["SO2_imp2"] = df["SO2"]

def imputar_por_tasa(grupo):
    grupo = grupo.set_index("fecha").copy()

    so2 = grupo["SO2"].copy()
    so2_omi = grupo["SO2_OMI"].copy()

    # Calcular la tasa de variación intermensual en SO2_OMI
    tasa = so2_omi.pct_change()

    # Imputación solo entre valores conocidos (no extrapolar)
    idx = so2.index
    for i in range(1, len(idx)-1):  # no miramos el primer ni último mes para evitar extrapolar
        if pd.isna(so2.iloc[i]):
            # Encontrar el último valor válido antes
            prev_idx = i - 1
            while prev_idx >= 0 and pd.isna(so2.iloc[prev_idx]):
                prev_idx -= 1

            # Encontrar el siguiente valor válido después
            next_idx = i + 1
            while next_idx < len(so2) and pd.isna(so2.iloc[next_idx]):
                next_idx += 1

            # Solo imputar si existe un valor anterior válido
            if prev_idx >= 0 and not pd.isna(so2.iloc[prev_idx]):
                # Aplicar tasa acumulada desde prev_idx hasta este mes
                valor = so2.iloc[prev_idx]
                for j in range(prev_idx+1, i+1):
                    if not pd.isna(tasa.iloc[j]):
                        valor *= (1 + tasa.iloc[j])
                    else:
                        # Si no hay tasa ese mes, no imputamos (queda NaN)
                        valor = np.nan
                        break
                so2.iloc[i] = valor

    grupo["SO2_imp2"] = so2
    return grupo.reset_index()

# Aplicamos por cada grid_id
df = df.groupby("grid_id", group_keys=False).apply(imputar_por_tasa)

# Orden final
df = df.sort_values(["grid_id", "fecha"])
```

```python
# 1. Agrupar por año y mes
df_grouped = df.groupby(["year", "month"], as_index=False)[["SO2_imp2", "SO2_OMI"]].mean()

# 2. Columna de fecha (primer día de cada mes)
df_grouped["fecha"] = pd.to_datetime(df_grouped[["year", "month"]].assign(day=1))

# 3. Crear figura y ejes
fig, ax1 = plt.subplots(figsize=(12,6))

# Ancho de barra: 20 días (para que se vea bien en Colab)
bar_width = pd.Timedelta(days=20)

# Eje 1 - Barras SO2
ax1.bar(df_grouped["fecha"], df_grouped["SO2_imp2"],
        width=bar_width, color="blue", label="SO2 extrapolado")
ax1.set_ylabel("SO2_imp2", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

# Eje 2 - Línea SO2_OMI
ax2 = ax1.twinx()
ax2.plot(df_grouped["fecha"], df_grouped["SO2_OMI"],
         color="red", marker="o", label="SO2_OMI")
ax2.set_ylabel("SO2_OMI", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# Título y formato
fig.suptitle("Comparación mensual SO2 imputación 2 (barra) vs SO2_OMI (línea)", fontsize=14)
fig.autofmt_xdate()

# Leyendas
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

plt.show()
```

Me gusta más esta imputación porque mantiene la estacionalidad en la serie de OMI.
